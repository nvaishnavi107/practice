```python
from pyspark.sql.functions import split, explode, col

data = [
    (1, "1,2,7,9,15"),
    (2, "12,9,18,10")
]

df=spark.createDataFrame(data,["id","values"])
df1=df.withColumn("num",explode(split(col("values"),",")))
df1.display()
```

```python
from pyspark.sql.functions import split, explode, col

data = [
    ("Pepporni", 0.5),
    ("Sausage", 0.7),
    ("Chicken",0.55),
    ("Extra cheese",0.4)
]
toppings=spark.createDataFrame(data,["topping_name","ingredients_cost"])
toppings.display()
```

```python
# cross join 3 times
t1=toppings.alias("t1")
t2=toppings.alias("t2")
t3=toppings.alias("t3")

from pyspark.sql.functions import concat_ws
df_final=t1.crossJoin(t2).crossJoin(t3).filter((col("t1.topping_name") < col("t2.topping_name")) & (col("t2.topping_name") < col("t3.topping_name")) ).withColumn("pizza",concat_ws(",",col("t1.topping_name"),col("t2.topping_name"),col("t3.topping_name"))).withColumn("cost_new",col("t1.ingredients_cost")+col("t2.ingredients_cost")+col("t3.ingredients_cost")).select("pizza","cost_new")


df_final.display()
```
